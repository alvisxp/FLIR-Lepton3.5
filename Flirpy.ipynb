### Flirpy 

```python
from flirpy.camera.lepton import Lepton
import cv2
import matplotlib.pyplot as plt
```

```python
cam = Lepton()
image = cam.grab()
```

```python
import numpy as np
print((image/100 - 273))
temp_cel = np.array((image/100 -273), copy = 'true')
np.savetxt('TempData.xls', image/100 - 273, fmt = '%.2f', delimiter = '\t'  )
```

```python
plt.imshow(image, cmap = 'plasma')
```

```python
cam.close()
```

```python
print('Max Temperature is: ', temp_cel.max(), "degree celcius")
print('Min Temperature is: ', temp_cel.min(), "degree celcius")
print('Average Temperature is:', temp_cel.mean(), "degree celcius")
```
